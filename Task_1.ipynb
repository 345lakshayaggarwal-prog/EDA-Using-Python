**MISSING VALUE HANDLING**

```python
df.isna().mean()*100     # % of missing values per column
df.isna().sum()          # Count missing per column
```

```python
# Drop columns with >70% missing values
df = df.dropna(axis=1, thresh=int(0.3*len(df)))
```

```python
# Drop rows with any missing values
df = df.dropna()
```

```python
# Numerical: fill with mean
df["Price"].fillna(df["Price"].mean(), inplace=True)
df["Price"].fillna(df["Price"].median(), inplace=True)
df["Price"].fillna(df["Price"].mode()[0], inplace=True)
```

**DATA TYPE CORRECTION**

```python
# Converting Strings to Datetime
df["Price"] = pd.to_datetime(df["Price"], errors="coerce")
```

```python
# Remove commas and convert to numeric
df["Stock"] = pd.to_numeric(df["Stock"])
```

```python
# Remove currency symbols
df["Currecy"] = df["Currency"].str.replace("USD", "$", regex=True).astype(object)
```
